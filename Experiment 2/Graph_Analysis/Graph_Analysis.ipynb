```python
import matplotlib.pyplot as plt
array_sizes = [5000, 10000, 15000, 20000, 25000]
merge_sort_times = [0.001000, 0.002300, 0.003500, 0.004500, 0.006000]
quick_sort_times = [0.000500, 0.001000, 0.001700, 0.002000, 0.002300]

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(array_sizes, merge_sort_times, label='Merge Sort Time', marker='o', color='b')
plt.plot(array_sizes, quick_sort_times, label='Quick Sort Time', marker='s', color='r')

# Adding labels and title
plt.title('Merge Sort vs Quick Sort Time Complexity')
plt.xlabel('Array Size')
plt.ylabel('Time (seconds)')
plt.legend()

# Display grid and show plot
plt.grid(True)
plt.show()
```
