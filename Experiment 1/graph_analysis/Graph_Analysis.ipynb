```python
import matplotlib.pyplot as plt

sizes = [5000, 10000, 15000, 20000, 25000]
recursive_times = [0.111, 0.443, 0.993, 1.739, 2.723]
iterative_times = [0.025, 0.105, 0.228, 0.406, 0.64]


plt.figure(figsize=(10, 6))
plt.plot(sizes, recursive_times, marker='o', color='r', label='Recursive Insertion')
plt.plot(sizes, iterative_times, marker='o', color='b', label='Iterative Insertion')


plt.title("Comparison of Recursive vs Iterative Insertion in BST", fontsize=14)
plt.xlabel("Number of Elements", fontsize=12)
plt.ylabel("Time Taken (seconds)", fontsize=12)

plt.legend()


plt.grid(True)
plt.show()
```
